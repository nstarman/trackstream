### Functions

```python
def make_dts(ordered_data, dt0, *, N=6, vmin=0.01, axis=1, plot=False):
    ii = np.arange(len(ordered_data)-1)  # indices
    ds = np.linalg.norm(ordered_data[1:] - ordered_data[:-1], axis=axis)  # point-to-point distance

    Ds = np.convolve(ds, np.ones((N,))/N, mode='same')
    Ds[Ds < vmin] = vmin

    dts = np.insert(Ds, 0, values=dt0)
    
    if plot:
        fig, ax = plt.subplots()
        ax.scatter(ii, ds, label="p2p distance")
        ax.scatter(ii, Ds, label="convolved")
        ax.set_xlabel("index")
        ax.set_ylabel("distance")
        ax.set_title("point to point distance")

        plt.legend()
        plt.tight_layout()
        plt.show
    
    return dts
```

```python
def make_F(dt):
    F_comp = np.array([[1, dt],
                       [0, 1]])
    F = block_diag(F_comp, F_comp, F_comp)
    return F
# /def
```

```python
def make_Q(dt, dim=2, var=1, block_size=3):
    Q = Q_continuous_white_noise(dim=dim, dt=dt, spectral_density=var, block_size=block_size)
    return Q
# /def
```

```python
run_output = namedtuple("run_output", field_names=["Xs", "Ps", "Fs", "Qs"])

def run_with_updatestep(data, dts, x0, P, R, H, u=0., B=0, alpha=1., *, qkw={}):
    x = x0

    Xs, Ps, Fs, Qs = [], [], [], []
    for i, z in enumerate(data):
        # F, Q
        F = make_F(dts[i])
        Q = make_Q(dts[i], **qkw)

        # predict
        x, P = predict(x, P=P, F=F, Q=Q, u=u, B=B, alpha=alpha)

        # update
        x, P, y, K, S, loglik = update(x, P=P, z=z, R=R, H=H, return_all=True)

        Xs.append(x)
        Ps.append(P)
        Fs.append(F)
        Qs.append(Q)

    Xs, Ps = np.array(Xs), np.array(Ps)
    Fs, Qs = np.array(Fs), np.array(Qs)

    # smooth
    sXs, sPs, sFs, sQs = rts_smoother(Xs, Ps, Fs, Qs)
    
    output = run_output(Xs, Ps, Fs, Qs)
    smooth = run_output(sXs, sPs, sFs, sQs)
    
    return output, smooth
```

```python
def plot_path(data, path, cov=None, true_path=None, *, cov_alpha=0.5, is_ordered=False):
    fig, axs = plt.subplots(1, 3, figsize=(15, 4))

    axs[0].set_title("x, y")
    axs[0].scatter(data.x, data.y)
    if is_ordered:
        axs[0].scatter(data.x[0], data.y[0], c="y", s=100)
    if true_path is not None:
        axs[0].plot(true_path.x, true_path.y, c="g")
    if cov is not None:
        plt.sca(axs[0])
        for i, p in enumerate(cov):
            P = np.array([[p[0, 0], p[2, 0]], 
                          [p[0, 2], p[2, 2]]])
            mean = (path.x[i], path.y[i])
            plot_covariance(mean, cov=P, fc='gray', std=3, alpha=cov_alpha)
        axs[0].set_aspect("auto")
    axs[0].plot(path.x, path.y, c="k")

    axs[1].set_title("x, z")
    axs[1].scatter(data.x, data.z)
    if is_ordered:
        axs[1].scatter(data.x[0], data.z[0], c="y", s=100)
    if true_path is not None:
        axs[1].plot(true_path.x, true_path.z, c="g")
    if cov is not None:
        plt.sca(axs[1])
        for i, p in enumerate(cov):
            P = np.array([[p[0, 0], p[4, 0]], 
                          [p[0, 4], p[4, 4]]])
            mean = (path.x[i], path.z[i])
            plot_covariance(mean, cov=P, fc='gray', std=3, alpha=cov_alpha)
        axs[1].set_aspect("auto")
    axs[1].plot(path.x, path.z, c="k")

    axs[2].set_title("y, z")
    axs[2].scatter(data.y, data.z)
    if is_ordered:
        axs[2].scatter(data.y[0], data.z[0], c="y", s=100)
    if true_path is not None:
        axs[2].plot(true_path.y, true_path.z, c="g")
    if cov is not None:
        plt.sca(axs[2])
        for i, p in enumerate(cov):
            P = np.array([[p[2, 2], p[4, 2]], 
                          [p[2, 4], p[4, 4]]])
            mean = (path.y[i], path.z[i])
            plot_covariance(mean, cov=P, fc='gray', std=3, alpha=cov_alpha)
        axs[2].set_aspect("auto")
    axs[2].plot(path.y, path.z, c="k")

    plt.tight_layout()
    
    return fig, axs
```

### Kalman Filter

```python
orb_kf = KalmanFilter(
    dim_x=6,  # x, dx, y, dy, z, dz
    dim_z=3,  # (x, y, z)
)
```

#### State transition function

For this function we implement a time dependency.

```python
orb_dts = make_dts(orb_arr, dt0=0.5, N=6, axis=1, plot=True)
orb_dt = orb_dts[0]
```

```python
orb_kf.F = make_F(orb_dt)
orb_kf.F
```

#### Process noise matrix

```python
orb_kf.Q = make_Q(dim=2, dt=orb_dt, var=0.01, block_size=3)

orb_kf.Q
```

#### Measurement function

```python
# measurement components
# only measure the position 
h = np.array([[1, 0],
              [0, 0]])

# full matrix is for all components
# and reduce down to `dim_z` of Kalman Filter, skipping velocity rows
orb_kf.H = block_diag(h, h, h)[::2]

orb_kf.H
```

#### Measurement noise matrix

```python
orb_kf.R = np.diag([
    0.05,  # error in x
    0.05,  # error in y
    0.003  # error in z
])
```

#### Initial conditions

Set the initial conditions to the first data point.

```python
x = orb_arr[0]  # fist point
v = [0, 0, 0]

orb_kf.x = np.array([x, v]).T.flatten()  # intersperse velocity data
orb_kf.x
```

#### State Covariance

```python
p = np.array([[0.0001, 0],  # very certain
              [0, 1]])  # very uncertain

P = block_diag(p, p, p)

orb_kf.P = P
orb_kf.P
```

#### Run

```python
orb_output, orb_smooth = run_with_updatestep(orb_arr, orb_dts, x0=orb_kf.x,
                                             P=orb_kf.P, R=orb_kf.R,
                                             H=orb_kf.H, u=0, B=0, alpha=1.,
                                             qkw=dict(dim=2, var=0.01, block_size=3))
```
